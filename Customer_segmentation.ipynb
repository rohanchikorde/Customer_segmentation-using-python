```python
import pandas as pd
```

```python
df_offers = pd.read_excel("C:\Users\dell\Documents\python project\Python project- customer segmentation\WineKMC.xlsx", sheetname =0)
```

```python
df_offers.head()  ## to check first few lines of data set
```

```python
df_offers.columns = ['offer_id', 'campaign', 'varietal', 'min_qty', 'discount',
                    'origin', 'past_peak']
```

```python
df_offers.head()
```

```python
df_transactions = pd.read_excel("C:\Users\dell\Documents\python project\Python project- customer segmentation\WineKMC.xlsx", sheetname = 1)
```

```python
## naming the column names
df_transactions.columns = ['customer_name', 'offer_id']
```

```python
df_transactions.head()
```

```python
## adding new column 'n'
```

```python
df_transactions['n'] = 1
df_transactions.head()
```

```python
## now we have 2 table -- offers and transaction table
## join both table
df = pd.merge(df_offers, df_transactions)
```

```python
## creating a pivot table which will give us the number of times each customer
## responded to a given offer

matrix = df.pivot_table(index =['customer_name'], columns = ['offer_id'], 
                        values = 'n')
```

```python
matrix
```

As data looks little tidy, fill NA with zero and make the index into columns

```python
matrix = matrix.fillna(0).reset_index()
```

```python
matrix.head()
```

```python
x_cols = matrix.columns[1:]
```

```python
## now we will k means for clustering
## install scikit-learn library and then import

from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters =5)
```

```python
## slicing the matrix so we only include 0/1 indicator columns in the clustering
```

```python
matrix['cluster'] = cluster.fit_predict(matrix[matrix.columns[2:]])
```

```python
matrix.cluster.value_counts()
```

Visualizing the Cluster using PCA Algorithm

```python
## import PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components= 2) 
## converting multidimensional 
## dataset to 2 dimensional dataset
```

```python
matrix['x'] = pca.fit_transform(matrix[x_cols])[:,0]
```

```python
matrix['y'] = pca.fit_transform(matrix[x_cols])[:,1]
```

```python
matrix = matrix.reset_index()
```

```python
customer_clusters = matrix[['customer_name', 'cluster', 'x', 'y']]
```

```python
customer_clusters.head()
```

```python
df = pd.merge(df_transactions, customer_clusters)
```

```python
df = pd.merge(df_offers,df)
```

```python
from ggplot import *
```

```python
ggplot(df, aes(x = 'x' , y='y', color = 'cluster')) + \
geom_point(size = 75) + \
ggtitle('Customers Grouped by Cluster')
```

```python
cluster_centers = pca.transform(cluster.cluster_centers_)
cluster_centers = pd.DataFrame(cluster_centers, columns=['x', 'y'])
cluster_centers['cluster'] = range(0, len(cluster_centers))

ggplot(df, aes(x='x', y='y', color='cluster')) + \
    geom_point(size=75) + \
    geom_point(cluster_centers, size=50) +\
    ggtitle("Customers Grouped by Cluster")
```
